```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout, Conv1D, MaxPooling1D
from keras.callbacks import EarlyStopping
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def load_and_preprocess_data(file_path):
    data = pd.read_csv(file_path)
    data['datetime'] = pd.to_datetime(data['datetime'])
    data = data.sort_values(by='datetime')
    data.set_index('datetime', inplace=True)
    return data

def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:(i + seq_length), 0])
        y.append(data[i + seq_length, 0])
    return np.array(X), np.array(y)

def build_model(input_shape):
    model = Sequential([
        Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape),
        MaxPooling1D(pool_size=2),
        LSTM(100, return_sequences=True),
        Dropout(0.2),
        LSTM(50, return_sequences=False),
        Dropout(0.2),
        Dense(50, activation='relu'),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

def train_and_evaluate_model(X_train, y_train, X_test, y_test):
    model = build_model((X_train.shape[1], 1))
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    history = model.fit(X_train, y_train, epochs=100, batch_size=32, 
                        validation_split=0.1, callbacks=[early_stopping], verbose=1)
    
    # Make predictions
    train_predict = model.predict(X_train)
    test_predict = model.predict(X_test)
    
    return model, history, train_predict, test_predict

def evaluate_model(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    return rmse, mae, r2

def plot_results(y_true, y_pred, title):
    plt.figure(figsize=(14, 7))
    plt.plot(y_true, color='blue', label='Actual PV Production (KW)')
    plt.plot(y_pred, color='red', label='Predicted PV Production (KW)')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('PV Production (KW)')
    plt.legend()
    plt.show()

# Main execution
file_paths = [
    "D:/Case Study/Baseline model/Dpt_1_2_PV.txt",
    "D:/Case Study/Baseline model/ENERPOS_PV.txt",
    "D:/Case Study/Baseline model/ESIROI_PV.txt"
]

for file_path in file_paths:
    print(f"Processing {file_path}")
    
    # Load and preprocess data
    data = load_and_preprocess_data(file_path)
    
    # Normalize the data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data[['Prod_kW']])
    
    # Prepare sequences
    seq_length = 24
    X, y = create_sequences(scaled_data, seq_length)
    
    # Split the data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # Reshape input for LSTM [samples, time steps, features]
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
    
    # Train and evaluate model
    model, history, train_predict, test_predict = train_and_evaluate_model(X_train, y_train, X_test, y_test)
    
    # Inverse transform predictions
    train_predict = scaler.inverse_transform(train_predict)
    test_predict = scaler.inverse_transform(test_predict)
    y_train_inv = scaler.inverse_transform(y_train.reshape(-1, 1))
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
    
    # Evaluate model
    rmse, mae, r2 = evaluate_model(y_test_inv, test_predict)
    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")
    print(f"R2 Score: {r2}")
    
    # Plot results
    plot_results(y_test_inv, test_predict, f'Actual vs Predicted PV Production - {file_path.split("/")[-1]}')
    
    # Get dates for test set
    test_dates = data.index[-len(y_test_inv):]
    
#     # Plot results
#     plot_results(y_test_inv, test_predict, test_dates, f'Actual vs Predicted PV Production - {file_path.split("/")[-1]}')
    
    # Save results
    results_df = pd.DataFrame({
        'Datetime': test_dates,
        'Actual': y_test_inv.flatten(),
        'Predicted': test_predict.flatten()
    })
    results_df.to_csv(f'results_{file_path.split("/")[-1]}.csv', index=False)
    print(f"Results saved to results_{file_path.split('/')[-1]}.csv")
    print("\n")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming `df` contains 'datetime', 'actual' and 'predicted' columns
df= pd.read_csv("results_Dpt_1_2_PV.txt.csv")

# Ensure the 'datetime' column is in datetime format
df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')

# Aggregate daily production
df['date'] = df['Datetime'].dt.date
daily_actual = df.groupby('date')['Actual'].sum()
daily_predicted = df.groupby('date')['Predicted'].sum()

# Plot daily production comparison
daily_actual.plot(label='Actual Daily Production', figsize=(10,6))
daily_predicted.plot(label='Predicted Daily Production')
plt.title('Predicted PV Production')
plt.xlabel('Date')
plt.ylabel('Energy Production (kW)')
plt.legend()
plt.show()

# Aggregate weekly production
df['week'] = df['Datetime'].dt.isocalendar().week
weekly_actual = df.groupby('week')['Actual'].sum()
weekly_predicted = df.groupby('week')['Predicted'].sum()

# # Plot weekly production comparison
weekly_actual.plot(label='Actual Weekly Production', figsize=(10,6), marker='o')
weekly_predicted.plot(label='Predicted Weekly Production')
plt.title('Weekly Predicted PV Production')
plt.xlabel('Week')
plt.ylabel('Energy Production (kW)')
plt.legend()
plt.show()

# # Aggregate monthly production
df['month'] = df['Datetime'].dt.month
monthly_actual = df.groupby('month')['Actual'].sum()
monthly_predicted = df.groupby('month')['Predicted'].sum()

# # Plot monthly production comparison
monthly_actual.plot(label='Actual Monthly Production', figsize=(10,6), marker='o')
monthly_predicted.plot(label='Predicted Monthly Production')
plt.title('Monthly Predicted PV Production')
plt.xlabel('Month')
plt.ylabel('Energy Production (kW)')
plt.legend()
plt.show()
```
